**GC Contents in DNA**

*   GC Content (guanine-cytosine content) is the percentage of nitrogenos bases in a DNA or an RNA molecule that are either Guanine (G) or Cytosine (C)

**Uses of GC Content**


*   A Higher GC-content level indicates a relatively higher melting temperature.
*   A DNA molecule with higher GC-Content is more stable than DNA with lower GC-Content
*   In polymerase chain reaction (PCR) experiments, GC-content of primers is often used to predict their annealing temperature to the template DNA.
*   A DNA with high GC content will be more challenging to perform PCR amplication due to difficulty in designing a primer long enough to provide great specifity

# AT Contents in DNA

**What is the AT Content of a DNA?**


*   AT Content is the percentage of nitrogenous bases in a DNA or RNA molecule that is either Adenine (A) or Thymine (T)
*   AT Base pairs only have TWO hydrogen bonds, while GC has THREE hydrogen bonds. That is the reason why a DNA with a higher GC-content tends to be more stabile



```python
# METHOD 1
# Create an iteration function to get AT Content
def at_content(seq):
  at_list = [n for n in seq.upper() if n in 'AT']
  return len(at_list) / len(seq) * 100

at_content(dna_seq)
```

# Melting Point of DNA

Melting temperature of a DNA refers to the temperature at which 50% of DNA in a sample has denatured from double-stranded DNA to single-stranded DNA (dsDNA to ssDNA)

DNA sequences with higher GC content tends to have a higher melting point due to its higher hydrogen bonds
image.png

**Several Ways to get Melting Point of DNA**


*   Tm_Wallace
>1. The Wallace Rule also known as "The rule of thumb"
>2. The formula is Tm = 2(A+T) + 4(G+C)


*   Tm_GC
>1. Empirical formulas based on GC content.
>2. Salt and mismatch corrections can be included.

*   Tm_NN
>1. Calculation based on nearest neighbor thermodynamics.
>2. Several tables for DNA/DNA, DNA/RNA and RNA/RNA hybridizations are included.
>3. Correction for mismatches, dangling ends, salt concentration and other additives are available.
