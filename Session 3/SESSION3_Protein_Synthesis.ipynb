# DNA Replication

DNA-replication.png

DNA Replication is a process by which a cell makes a copy of its entire DNA molecule, which is essential for cell division and the passing on of genetic information from one generation of cells to the next.


During DNA replication, the two strands of the DNA double helix are separated and each strand serves as a template for the synthesis of a new complementary strand.

*   **Cytosine** Pairs with **Guanine**
*   **Adenine** Pairs with **Thymine**

````python
# DNA Replication

dna_seq.complement()
````

````python
# DNA Reverse Complement
# Baca nya dari kanan ke kiri
dna_seq.reverse_complement()
````

````python
# Reverse Complement with python syntax
dna_seq.complement()[::-1] # -1 step, artinya baca dari kanan ke kiri
````

# Protein Synthesis



**Transcription**

A process by which DNA is copied (transcribed) to a messenger RNA (mRNA), which carries the information needed for protein synthesis.








````python
# Create a DNA Sequence
dna_seq
````

````python
# Transcription : DNA -> mRNA
# mRNA = messenger RNA
mRNA = dna_seq.transcribe()
mRNA
````

````python
# Back Transcribe : mRNA -> DNA

dna_back = mRNA.back_transcribe()
dna_back
````

````python
# Proof that the result of back transcribe is the same as the initial DNA

dna_back == dna_seq
````

codon-table-1024x871.png



**Translation**

During translation, the mRNA sequence is decoded into a sequence of amino acids that are linked together by peptide bonds to form a polypeptide chain.



````python
# Translation METHOD 1
# From RNA -> Protein / Amino Acid

protein = mRNA.translate()
protein
````

````python
# Translation METHOD 2
# Direct translation from DNA -> Protein / Amino Acid

protein = dna_seq.translate()
protein
````

````python
# Translation METHOD 3
# DNA Transcribe -> mRNA, mRNA Translate -> Protein

dna_seq.transcribe().translate()
````

````python
# Translation with custom stop codon symbol

custom_protein = mRNA.translate(stop_symbol = "!")
custom_protein
````

````python
# Import seq1 and seq3 to convert amino acid's letter codes

from Bio.SeqUtils import seq1, seq3
````

````python
# Convert Amino Acid's 1 letter code to 3 letter code

protein_3_letter = seq3(protein)
protein_3_letter


#Ter is termination
````

````python
custom_protein_3_letter = seq3(custom_protein)
custom_protein_3_letter

#Seq3 hanya bisa membaca * sebagai stop codon
````

````python

# Convert Amino Acid's 3 letter code to 1 letter code

seq1(protein_3_letter)
````

**Parsing and Analyzing Proteins from PDB Files**

Reading PDB Files from


*   https://www.ncbi.nlm.nih.gov/Structure/pdb/6LU7
*   https://www.rcsb.org/search
*   Protein Data Bank

Then, analyzing protein structure. The layout of the structure object follows SMCRA Architecture:

*Structure -> Model -> Chain -> Residue -> Atom*


> **File Format:**
> *   PDB (Legacy) : PDBParser()
> *   CIF (Recent) : MMCIFParser()



````python
# Import PDB dari downloads folder
# Ada gambar folder di kiri, pencet upload, trus upload pdb filenya
````

````python
# Import File Parser
from Bio.PDB import PDBParser
````

````python
# Abbreviate PDBParser

pdb = PDBParser()
````

````python
# Upload 6LU7.pdb file, then
# Get covid's structure by parsing PDB File

structure = pdb.get_structure("6LU7", "/content/6lu7.pdb")
# sebelah kiri adalah nama, sebelah kanan adalah directory file

structure
````

````python
# Get models from structure

model = structure[0]
model
````

````python
# Check for Chains

for chain in model:
  print(f"Chain {chain} | ChainID {chain.id}")
````

````python
# Check for Residue


for chain in model:
  print(f"Chain {chain} | ChainID {chain.id}")

  for residue in chain:
    print(f"Residue {residue}")
````

````python
# Check for Atoms

for chain in model:
  print(f"Chain {chain} | ChainID {chain.id}")

  for residue in chain:
    print(f"Residue {residue}")

    for atom in residue:
      print(f"\t\tAtom {atom}")
````
