```python
# Import Seq
from Bio.Seq import Seq
```

```python
# Check Seq's Methods
dir(Seq)
```

```python
# Create a simple DNA Sequence
dna = Seq('CGATTCGAATT')
dna
```

```python
# Create a simple RNA Sequence
rna = Seq('CGAUUCGAAUU')
rna
```

```python
# Create a Protein Seq
protein = Seq('MIS')
protein
```

# **Seqence Manipulation**

*   Indexing & Slicing

```python
# Get the Length of Seq
len(dna)
```

```python
# Slicing
#dna = dna[1:-2]
dna[1:-2]
```

```python
# Reverse
dna[::-1]
```



*   Joining 2 Sequences



```python
# Join Seq
new_seq = Seq('CGAAATTTCGAT')
dna + new_seq
```



*   Find the number of Nucleotides and Codons in a sequence



```python
# Find the number of Nucleotides/Base within our seq
dna.count('C')
```

```python
# Find the number of Codons within our seq
dna.count('CGAT')
```



*   Finding the position / index of Nucleotiedes

```python
# Find the position / index of Nucleotide
dna.find('T')
```

```python
# Find the position / index of Nucleotide from the right
dna.rfind('T')
```

```python
# Find the position / index of Nucleotide using index
dna.index('T', 4)
```

```python
# Find the position / index of Nucleotide from the right using index
dna.rindex('T')
```

# **SubSequence**

```python
# Creating Main and Sub Sequence
main_seq = dna
sub_seq = Seq('CGAT')
```

```python
# Import nt_search from SeqUtils Package (To search subsequence in main sequence)
from Bio.SeqUtils import nt_search
```

```python
# Search the Subsequence from Main Sequence
nt_search(str(main_seq), str(sub_seq))
```

# **Plot Frequency for Nucleotides**


```python
# Install Matplotlib (Library for plotting)
import matplotlib
```

```python
# Import pyplot from matplotlib
from matplotlib import pyplot as plt
```

```python
# Import Counter function from collections package
from collections import Counter
```

```python
# Create Sequence and count Nucleotide Frequency
myseq = dna
myseq_freq = Counter(dna)
```

```python
# Plot the DNA sequence Nucleotide Frequency
plt.bar(myseq_freq.keys(), myseq_freq.values() )
plt.show()
```
