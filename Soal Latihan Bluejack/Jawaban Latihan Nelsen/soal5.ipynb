```python
import pandas as pd
```

```python
contentApa = SeqIO.read('./sequence.fasta', 'fasta')

sequencenyaini = contentApa.seq

sequencenyaini = sequencenyaini.transcribe().translate()
```

```python
protein_split = sequencenyaini.split('*')

protein_split = [str(se) for se in protein_split]

protein_split
```

```python
df = pd.DataFrame()
df["protein"] = protein_split
df["length"] = df["protein"].str.len()

df.head()
df.nlargest(10, "length")
```
